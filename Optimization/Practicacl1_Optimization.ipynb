### 2) Analytically compute the gradient of the function, $\nabla f(x)$, and compute the point $x^∗$ at which $\nabla f(x^∗) = 0$.

**Solution**

In order to comput the gradient of the function, firstly we differentiate the function with respect to each of the variables $ x_1$ and $x_2$. Which is the vector of its partial derivatives with respect to each variable. That is:

$\nabla f(x_1, x_2) = ( \frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2}) $

Partial derivative with respect to $x_1$:

$\frac{\partial}{\partial x_1} f = \frac{\partial}{\partial x_1} (x^2_1+x^2_2) = 2x_1$

Partial derivative with respect to $x_2$:

$\frac{\partial}{\partial x_2} f = \frac{\partial}{\partial x_2} (x^2_1+x^2_2) = 2x_2$

Final gradient:
$\nabla f(x) = \begin{pmatrix}
           2x_1 \\
           2x_2
  \end{pmatrix}$

Now to compute the critical points, we equal the gradient to 0.
$\nabla f(x^*) = 0  ⇒ \begin{pmatrix}
           x_1 \\
           x_2
  \end{pmatrix} = \begin{pmatrix}
           0 \\
           0
  \end{pmatrix}$

  (0,0) is the only critical point.

###3)
Let $d \in R^2$ be the perturbation around $x^∗$. The Taylor expansion, up to second order, of a function of several variables can be compactly expressed as
$f(x^∗ + d) ≈ f(x^∗) + d^T \nabla f(x^∗) + \frac{1}{2}d^T \nabla ^2f(x^∗) d$ (2)

Analyze the previous expression and be sure to understand the operations that are done at each of the terms.
Compute the Hessian matrix, $\nabla ^2f(x)$, at the point $x = x^∗$. You should obtain
$$
\nabla ^2 f(x^∗ ) = \begin{pmatrix}
           2 \ 0\\
           0 \ 2
  \end{pmatrix}
$$

The latter matrix is giving us information about the shape of the quadratic approximation at
$x = x^∗$ in a similar way as has been done for the one dimensional case.

For the one-dimensional case it is easy to check if we have a minimum, $f''(x^∗) > 0$, or a
maximum, $f''(x^∗) < 0$. For a higher dimensional problem we are sure that the quadratic approximation is convex and that we have a minimum if

$d^T \nabla ^2 f(x^∗) d >0 d \neq 0 $ (3)

We have a maximum if

$d^T \nabla ^2 f(x^∗) d < 0 d \neq 0$ (4)

The previous conditions can be verified by computing the eigenvalues of $\nabla ^2f(x^∗)$. If all eigenvalues are strictly positive, equation (3) is satisfied. If all eigenvalues are strictly negative, equation (4) is satisfied. **For this example which are the eigenvalues of the Hessian matrix? Do we have a minimum or a maximum at $x^∗$?**

**Solution:**

Computing the Hessian Matrix as it contains de second-order partial derivatives.

$$
\nabla^2 f(x) = \begin{pmatrix}
           \frac{\partial^2 f}{\partial x^2} & \frac{\partial^2 f}{\partial x \partial y} \\
           \frac{\partial^2 f}{\partial y \partial x} & \frac{\partial^2 f}{\partial y^2}
           \end{pmatrix} =
\begin{pmatrix}
          2 & 0 \\
          0 & 2
          \end{pmatrix}
$$

For this example the eigenvalues of the Hessian Matrix we need to solve the characteristic equation:

$det( A - \lambda I) = 0 ⇒ $

$$
\begin{pmatrix}
          2 & 0 \\
          0 & 2
          \end{pmatrix} - \lambda
\begin{pmatrix}
          1 & 0 \\
          0 & 1
          \end{pmatrix} =
\begin{pmatrix}
          2- \lambda & 0 \\
          0 & 2- \lambda
          \end{pmatrix}
$$
And now we compute the determinant to classify the critical point (0,0).

$$
det\begin{bmatrix}
          2- \lambda & 0 \\
          0 & 2- \lambda
          \end{bmatrix} = (2 - \lambda) (2 - \lambda) = (2 - \lambda) ^2
$$

Now solve for $ \lambda $

$$(2 - \lambda) ^2 = 0 ⇒ \lambda = 2$$

Thus, the **eigenvalues** of the matrix
$
A = \begin{pmatrix}
          2 & 0 \\
          0 & 2
          \end{pmatrix}
$ are
$\lambda_1 = 2$ and $\lambda_2 = 2$.

Now evaluate the Hessian Matrix at critical points in order to know if they are maximums or minimums.

Because the eigenvalues are positive ($ \lambda = 2 > 0$), we can say that the critical point **$x^* = (0,0)$ is a minimum**.

##### (b) Using brute force, search for those points $x^\sim$  within the previous range at which the value of $||\nabla f(x)||^2$ is strictly smaller than the value of its 8 neighbors. Our purpose here is to find the those points at which the gradient is small. We thus find all the “candidate” points that may be a minimum, a maximum or a saddle point!

```python
# Define a function to check if the value at (i, j) is smaller than its 8 neighbors
def is_local_min(matrix, i, j):
    # Get the value at (i, j)
    center_val = matrix[i, j]

    # Get the values of its 8 neighbors
    neighbors = [
        matrix[i-1, j-1], matrix[i-1, j], matrix[i-1, j+1],
        matrix[i, j-1],                   matrix[i, j+1],
        matrix[i+1, j-1], matrix[i+1, j], matrix[i+1, j+1]
    ]

    # Check if the value is strictly smaller than all its neighbors
    return all(center_val < neighbor for neighbor in neighbors)
```

```python
# Now let's find the points in the matrix

minims = []
for i in range(1,len(x2)-1):
    for j in range(1,len(x1)-1):
        v = ng[i][j]
        if is_local_min(ng, i, j):
            minims.append([i,j, x1[j], x2[i], v])

# Show the results found
print("  i   j \t x1 \t\t\t x2 \t\t ||∇f(x)||")
for m in minims:
    print(m)
```

```python
# We can check that these points are the minimum in their neighboorhood
for m in minims:
    i = m[0]
    j = m[1]
    small = ng[i-1:i+2, j-1:j+2]
    print(small, "\n")
```

#### 4. Compute the analytical expression of the Hessian of $f(x_1, x_2)$ and evaluate it at the values $x˜$ you have found. What kind of information is giving you the Hessian? Does it correspond to a minimum (the Hessian is positive definite)? To a maximum (the Hessian is negative definite)? Or may be a saddle point? You may use the eigvals function of Python to compute the eigenvalues of the Hessian matrix (no need to compute them analytically). Take into account that there may be several minima, maxima and saddle points for the function you are analyzing.

Calculating the Hessian analytically we get:

$$
\nabla^2 f(x) = \begin{pmatrix}
           \frac{\partial^2 f}{\partial x^2} & \frac{\partial^2 f}{\partial x \partial y} \\
           \frac{\partial^2 f}{\partial y \partial x} & \frac{\partial^2 f}{\partial y^2}
           \end{pmatrix} =
\begin{pmatrix}
          10x_1^4 - 25,2x_1^2 +8 & 1 \\
          1 & 48x_2^2 -8
          \end{pmatrix}
$$

```python
# Function to calculate the hessian matrix of a point

def hessian(X1, X2):
    m = np.ones((2,2))
    m[0, 0] = 10*X1**4 - 25.2*X1**2 + 8
    m[1, 1] = 48*X2**2 - 8
    return m

print("Example:")
print(hessian(0,0))
```

```python
# Let's analyse the Hessian in each candidate point
for m in minims:
    X1 = m[2]
    X2 = m[3]
    h = hessian(X1, X2)
    evals = np.linalg.eigvals(h)

    print("x = ({:.3f}, {:.3f})".format(X1, X2))
    print("Hessian matrix:")
    print(h)
    print("Eigenvalues:", evals)

    if (np.all(evals > 0)):
        print('-> Local minimum.')
        m.append("Local minimum")
    elif (np.all(evals < 0)):
        print('-> Local maximum.')
        m.append("Local maximum")
    elif (evals[0] * evals[1] < 0):
        print('-> Saddle point.')
        m.append("Saddle point")
    else:
        print("Undefined")
        m.append("Undefined")
    print()
```
